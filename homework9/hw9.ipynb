# 2 Plotting a Function
Goal: Write a **function** that will plot $f(x) = x^2$ and $g(x) = \sqrt{x}$ on the same plot and return the $f(x)$ and $g(x)$ values as separate lists. Your plot needs to have the following:
- An appropriate figure size (`figsize=(10,8)` is recommended)
- Labeled x and y axis
- A title
- Different colors and linestyles for each function
- A legend
- Your domain should be $x \in [0, 10]$. *Hint: Use `np.linspace` to set a range of values for your functions.*

Your function should return:
- A list of $f(x)$ values
- A list of $g(x)$ values.

```python
# Your code here
import numpy as np
import matplotlib.pyplot as plt

def plot_functions():
    # defining the domain for x
    x = np.linspace(0, 10) # domain of x
    
    # defining the functions
    f_x = x**2
    g_x = np.sqrt(x)
    
    # Plot code from previous but new information
    plt.figure(figsize=(10, 8))
    
    # Plot f(x) = x^2
    plt.plot(x, f_x, color='blue', linestyle='-', label=r'$f(x) = x^2$')
    
    # Plot g(x) = sqrt(x)
    plt.plot(x, g_x, color='green', linestyle='--', label=r'$g(x) = \sqrt{x}$')
    
    #title and labels
    plt.title("Plot of $f(x) = x^2$ and $g(x) = \sqrt{x}$", fontsize=16)
    plt.xlabel("x", fontsize=14)
    plt.ylabel("y", fontsize=14)
    
    # the legend
    plt.legend()
    
    # showing grid
    plt.grid(True)
    
    # showing/displaying the plot
    plt.show()
    
    # return the function values as lists
    return f_x.tolist(), g_x.tolist()

# Call the function and store the returned values
f_values, g_values = plot_functions()
```

# 3 Introduction to Subplots
Goal: Write a **module** (name it whatever you want, as long as it is professional) that contains two functions that do the following: 

Function 1: 
- Write a **function** that will plot two subplots side-by-side (horizontal).
- The **left** subplot will be: $h(x) = \cos(x)$.
- The **right** subplot will be $k(x) = \sin(x)$.
- Include a title, axis labels and a domain of $x \in [0, 2 \pi]$ (this should NOT be hardcoded)

Function 2:
- Exactly the same as your first function, **EXCEPT** the subplots will be on top of each other (vertical). 

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_subplots_horizontal():
    #the domain for x
    x = np.linspace(0, 2 * np.pi, 100)  # domain x ∈ [0, 2π]
    
    # the functions h(x) = cos(x) and k(x) = sin(x)
    h_x = np.cos(x)
    k_x = np.sin(x)
    
    # subplots
    fig, axs = plt.subplots(1, 2, figsize=(12, 6))  # 1 row, 2 columns
    
    # Plot h(x) = cos(x) in the left subplot
    axs[0].plot(x, h_x, color='blue', label=r'$h(x) = \cos(x)$')
    axs[0].set_title(r'$h(x) = \cos(x)$', fontsize=14)
    axs[0].set_xlabel("x", fontsize=12)
    axs[0].set_ylabel("h(x)", fontsize=12)
    axs[0].grid(True)
    axs[0].legend(loc='upper right')

    # Plot k(x) = sin(x) in the right subplot
    axs[1].plot(x, k_x, color='red', label=r'$k(x) = \sin(x)$')
    axs[1].set_title(r'$k(x) = \sin(x)$', fontsize=14)
    axs[1].set_xlabel("x", fontsize=12)
    axs[1].set_ylabel("k(x)", fontsize=12)
    axs[1].grid(True)
    axs[1].legend(loc='upper right')
    
    #layout
    plt.tight_layout()
    
    # show/display
    plt.show()

def plot_subplots_vertical():
    # domain for x
    x = np.linspace(0, 2 * np.pi, 100)  # domain x ∈ [0, 2π]
    # functions h(x) = cos(x) and k(x) = sin(x)
    h_x = np.cos(x)
    k_x = np.sin(x)
    
    # subplots 
    fig, axs = plt.subplots(2, 1, figsize=(6, 12))  # 2 rows, 1 column
    
    # Plot h(x) = cos(x) in the top subplot
    axs[0].plot(x, h_x, color='blue', label=r'$h(x) = \cos(x)$')
    axs[0].set_title(r'$h(x) = \cos(x)$', fontsize=14)
    axs[0].set_xlabel("x", fontsize=12)
    axs[0].set_ylabel("h(x)", fontsize=12)
    axs[0].grid(True)
    axs[0].legend(loc='upper right')

    # Plot k(x) = sin(x) in the bottom subplot
    axs[1].plot(x, k_x, color='red', label=r'$k(x) = \sin(x)$')
    axs[1].set_title(r'$k(x) = \sin(x)$', fontsize=14)
    axs[1].set_xlabel("x", fontsize=12)
    axs[1].set_ylabel("k(x)", fontsize=12)
    axs[1].grid(True)
    axs[1].legend(loc='upper right')
    
    #layout
    plt.tight_layout()
    #show/display
    plt.show()
```
